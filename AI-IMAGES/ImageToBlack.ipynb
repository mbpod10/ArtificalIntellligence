```python
import numpy as np
from PIL import Image, ImageFilter
```

```python
# Load the image and convert it to grayscale
img = Image.open('./AI-IMAGES/gen_images/blackBeard.png').convert('L')

# Convert the image to a NumPy array
pixels = np.array(img)

# Set a threshold value
threshold = 128

# Set all pixel values above the threshold to 255 (white) and all others to 0 (black)
pixels[pixels > threshold] = 255
pixels[pixels <= threshold] = 0

# Convert the NumPy array back to an image
img = Image.fromarray(pixels)

# Display the modified image
display(Image.fromarray(np.array(img)))
```

```python
# Load the image into a NumPy array
img = np.array(Image.open('./AI-IMAGES/gen_images/blackBeard.png'))

# Convert the image to grayscale
img = Image.fromarray(img).convert('L')

# Apply a sharpening filter to the image
img_enhanced = img.filter(ImageFilter.SHARPEN)

# Display the original and enhanced images side by side
display(Image.fromarray(np.array(img)))
display(Image.fromarray(np.array(img_enhanced)))
```
